```python
# Dependencies and Setup
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as sts
import os
import numpy as np
%matplotlib inline
```

```python
# Study data files
mouse_metadata_path = os.path.join("Mouse_metadata.csv")
study_results_path = os.path.join("Study_results.csv")
```

```python
# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
```

```python
# Combine the data into a single dataset
df = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
df = df.rename(columns={"Tumor Volume (mm3)":"tumor_volume"})
# Display the data table for preview
df.head()
```

```python
# Checking the number of mice.
len(df["Mouse ID"].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
# Optional: Get all the data for the duplicate mouse ID.
#Check for duplicates and drop
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_df=df.drop_duplicates()
# Optional: Get all the data for the duplicate mouse ID.
```

```python
# Checking the number of mice in the clean DataFrame.
len(clean_df["Mouse ID"].unique())
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
```

```python
regimen_volume_clean_groupby=clean_df[["Drug Regimen", "tumor_volume"]].groupby("Drug Regimen")
tumor_volume_mean=regimen_volume_clean_groupby["tumor_volume"].mean()
tumor_volume_median=regimen_volume_clean_groupby["tumor_volume"].median()
tumor_volume_variance=regimen_volume_clean_groupby["tumor_volume"].var()
tumor_volume_standard_deviation=regimen_volume_clean_groupby["tumor_volume"].std()
tumor_volume_SEM=regimen_volume_clean_groupby["tumor_volume"].sem()
```

```python
drug_regimen_df=pd.DataFrame({"Mean":tumor_volume_mean,"Median":tumor_volume_median,"Standard Deviation":tumor_volume_standard_deviation,"Variance":tumor_volume_variance,"SEM":tumor_volume_SEM})
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
drug_regimen_agg_df=regimen_volume_clean_groupby.agg({"tumor_volume":["mean","median", "std", "var", "sem"]})
drug_regimen_agg_df
```

```python
# Calculate the IQR and quantitatively determine if there are any potential outliers.
tumor_volume=df2["tumor_volume"]
#Characterize the data
plt.hist(tumor_volume)
plt.show()
print(sts.normaltest(tumor_volume))
#According to the appearance of the histogram and the pvalue the dataset is not "normal"
quartiles=tumor_volume.quantile([.25,.5,.75])
lowerq=quartiles[0.25]
upperq=quartiles[.75]
iqr=upperq - lowerq
lower_bound=lowerq-(1.5*iqr)
upper_bound=upperq+(1.5*iqr)
print(f"The IQR is: {iqr}")
outliers=df2.loc[(df2['tumor_volume'] < lower_bound) | (df2['tumor_volume'] > upper_bound)]
print(outliers)

    
```
